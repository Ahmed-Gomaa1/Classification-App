```python
from tkinter import *
from tkinter import filedialog as fd
import tkinter as tk
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from tkinter import messagebox
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.svm import SVR
```

```python
root = Tk()
root.geometry('1200x650')
root.title('Classification App')

def get_Model1():
    global Model
    Model=radiobutton_variable.get()
    windos=Tk()
    windos.geometry('300x300')
    windos.title('KNeighborsClassifier Attributes')
    def getData1():
        global N_neighbors
        try:
            N_neighbors=int(En1.get())      
        except:
            messagebox.showerror("Error","Please Enter Valid Number For N_neighbors") 
        windos.destroy()
    lb1=Label(windos, text="Number Of N_neighbors", font=("Courier 10 bold"))
    lb1.place(x=70,y=90)
    En1 = Entry(windos)
    En1.place(x=90,y=110)       
    b1 = Button(windos, text="Done", command=getData1,bg="green",fg="white")
    b1.place(x=135,y=150)
    tk.mainloop()    

def get_Model2():
    global Model
    Model=radiobutton_variable.get()
    windos=Tk()
    windos.geometry('300x300')
    windos.title('DecisionTreeClassifier Attributes')
    def getData2():
        global Max_depth
        try:
            Max_depth=int(En1.get())
        except:
            messagebox.showerror("Error","Please Enter Valid Number For Max_depth ")
        windos.destroy()
    lb1=Label(windos, text="Number Of Max_depth", font=("Courier 10 bold"))
    lb1.place(x=70,y=90)
    En1 = Entry(windos)
    En1.place(x=90,y=110)       
    b1 = Button(windos, text="Done", command=getData2,bg="green",fg="white")
    b1.place(x=135,y=150)
    tk.mainloop()

def get_Model3():
    global Model
    Model=radiobutton_variable.get()
    windos=Tk()
    windos.geometry('300x300')
    windos.title('Linear Regression Attributes')
    def getData3():
        global N_estimators,RMax_depth
        try:
            N_estimators=int(En1.get())
        except:
            messagebox.showerror("Error","Please Enter Valid Number For N_estimators ")
        try:
            RMax_depth=int(En3.get())
        except:
            messagebox.showerror("Error","Please Enter Valid Number For RMax_depth ")
        windos.destroy()
            
    lb1=Label(windos, text="Number Of N_estimators", font=("Courier 10 bold"))
    lb1.place(x=70,y=50)
    En1 = Entry(windos)
    En1.place(x=70,y=70)
    lb2=Label(windos, text="Number Of Max_depth", font=("Courier 10 bold"))
    lb2.place(x=70,y=90)
    En3 = Entry(windos)
    En3.place(x=70,y=110)
    b1 = Button(windos, text="Done", command=getData3,bg="green",fg="white")
    b1.place(x=70,y=150)
    tk.mainloop()
    
def get_Model4():
    global Model
    Model=radiobutton_variable.get()
    windos=Tk()    
    windos.geometry('300x300')
    windos.title('SVC Attributes')
    def getData4():
        global Kernel,valuesButton
        valuesButton=int(v.get())
        if valuesButton==1:
            Kernel='linear'
        if valuesButton==2:
            Kernel='poly'
        if valuesButton==3:
            Kernel='rbf'
        windos.destroy()
    lb1=Label(windos, text="Kernel", font=("Courier 10 bold"))
    lb1.place(x=70,y=50)
    v = StringVar(windos, "1") 
    Radiobutton(windos, text='Linear', variable=v, value='1').place(x=70,y=70)
    Radiobutton(windos, text='Polynomial', variable=v, value='2').place(x=70,y=90)
    Radiobutton(windos, text='RBF', variable=v, value='3').place(x=70,y=110)
    b1 = Button(windos, text="Done", command=getData4,bg="green",fg="white")
    b1.place(x=125,y=150)
    tk.mainloop()
    
def callback():
    global data,columns
    data= fd.askopenfilename()
    data=pd.read_csv(data)
    columns=list(data.columns)

def get_value1():
    global n_features
    try:
        n_features=int(E1.get())
        if n_features > len(columns) or n_features<=1:
            messagebox.showerror("Error","Please Enter Number between 0 and Number of Features ")
    except:
           messagebox.showerror("Error","Please Enter Valid Number For Selected Features ")
        
def get_value2():
    global Text_size
    try:
        Text_size=float(E2.get())
    except:
        messagebox.showerror("Error","Please Enter Valid Float For Text_size ")
    if Text_size > 1:
        messagebox.showerror("Error","Please Enter Text_size between 0 and 1 ")
    else:
        Text_size=float(E2.get())
        
def Feature():
    global Target,x,y
    try:
        x=data.iloc[:,:-1]
        x=pd.get_dummies(x)
        col=(x.columns).copy()
        x=x.values
        y=data.iloc[:,-1]
        le= LabelEncoder()
        y = le.fit_transform(y)
        estimator = DecisionTreeClassifier()
        get_value1()
        selector = RFE(estimator, n_features_to_select=n_features,step=1)
        selector = selector.fit(x, y)
        Target=columns[-1]
        features=np.array(col)
        xc=features[selector.support_]
        x = data[xc].values
    except:
        messagebox.showerror("Error","Please Upload Your Dataset First")
        
        
         
def fit():
    Feature()
    get_value2()
    global knn,dt,rf,svm,checkfit,x_test , y_test,checkfit
    checkfit=0
    x_train , x_test , y_train , y_test = train_test_split(x,y, test_size= Text_size, random_state =30)
    if Model==0:
            messagebox.showerror("Error","Please Choose Model First")
    if Model==1:
            knn =KNeighborsClassifier(n_neighbors=N_neighbors)
            knn.fit(x_train,y_train)
            checkfit=1
    if Model==2:
            dt =DecisionTreeClassifier(max_depth=Max_depth)
            dt.fit(x_train,y_train)
            checkfit=1
    if Model==3:
            rf=RandomForestClassifier(max_depth=RMax_depth,n_estimators=N_estimators)
            rf.fit(x_train,y_train)
            checkfit=1
    if Model==4:
            svm =SVC(kernel=Kernel)
            svm.fit(x_train,y_train)
            checkfit=1
            
def testdata():
    global pred,label9
    if checkfit==0:
        messagebox.showerror("Error","Please Fit Model First")
    else :
        if Model==1:
            pred= knn.predict(x_test)
        if Model==2:
            pred= dt.predict(x_test)
        if Model==3:
            pred= rf.predict(x_test)
        if Model==4:
            pred= svm.predict(x_test)
            
def result():
        cm = confusion_matrix(y_test,pred)
        label1=Label(root, text='confusion_matrix is', font=("Courier 10 bold"))
        label1.place(x=80,y=425)
        label2=Label(root, text=cm, font=("Courier 10 bold"))
        label2.place(x=220,y=450)
        label4=Label(root, text='classification_report is', font=("Courier 10 bold"))
        label4.place(x=420,y=375)
        target=[]
        for i in set(data[Target]):
            target.append(str(i))
        label5=Label(root, text=classification_report(y_test, pred, target_names=target), font=("Courier 10 bold"))
        label5.place(x=470,y=400)

        
dst1 = Button(root, text="Click to Open File", command=callback,bg="green",fg="white")
dst1.place(x=320,y=10)
label6=Label(root, text="Number Of Selected Features For Model", font=("Courier 10 bold"))
label6.place(x=225,y=50)
E1 = Entry(root)
E1.place(x=310,y=70)

label7=Label(root, text="Define Test Size", font=("Courier 10 bold"))
label7.place(x=310,y=110)
E2 = Entry(root)
E2.place(x=310,y=130)

lbl = Label(root, text = "Choose Model", font=("Helvetica", 12), fg = 'white' , bg='black')
lbl.place(x=650,y=40)
radiobutton_variable = IntVar()
Radiobutton(root, text="KNeighborsClassifier", variable = radiobutton_variable, value = 1,command=get_Model1).place(x=660,y=70)
Radiobutton(root, text="DecisionTreeClassifier",  variable = radiobutton_variable, value = 2,command=get_Model2).place(x=660,y=100)
Radiobutton(root, text="RandomForestClassifier", variable = radiobutton_variable, value = 3,command=get_Model3).place(x=660,y=130)
Radiobutton(root, text="SupportVectorClassifier",  variable = radiobutton_variable, value = 4,command=get_Model4).place(x=660,y=160)

dst2 = Button(root, text="fit ", command=fit,font=("Helvetica", 12), fg = 'white' , bg='black')
dst2.place(x=400,y=180)
dst3 = Button(root, text="test", command=testdata,font=("Helvetica", 12), fg = 'white' , bg='black')
dst3.place(x=320,y=180)
dst4 = Button(root, text="show result", command=result,font=("Helvetica", 12), fg = 'white' , bg='black')
dst4.place(x=329,y=230)

tk.mainloop()
```
